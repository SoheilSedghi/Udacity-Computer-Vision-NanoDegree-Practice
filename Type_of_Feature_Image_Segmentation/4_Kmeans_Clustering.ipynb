# Kmeans Clustering

## Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## read image

```python
bgr = cv2.imread("20221214_203652.jpg")
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
plt.figure( figsize = (15,10))
plt.imshow(rgb)
```

## reshape for clustering

```python
print( "shape of rgb is:" ,rgb.shape)
pixel_val = np.float32(rgb.reshape((-1,3)))
print("after reshape:",pixel_val.shape)
```

## Kmeans clustering

```python
# stop criteria
criteria = (cv2.TERM_CRITERIA_EPS+ cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)

k= 2
retval, labels, centers = cv2.kmeans(pixel_val, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS )
centers = np.uint8(centers)
segmented_image = centers[labels.flatten()].reshape((rgb.shape))
plt.figure(figsize = (15,10))
plt.imshow(segmented_image, )
```
