# Import libraries

```python
import os
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

# Find image

```python
files = os.listdir("./")
plt.figure(figsize=(20,20))
for file in files:
    if file == "up.jpg":
        img = cv2.imread(file)
        plt.imshow(img)
        break
```

# Get RGP Mask

```python
RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
r = RGB[:, :, 0]
g = RGB[:, :, 1]
b = RGB[:, :, 2]
plt.figure(figsize=(20,15))
plt.subplot(1,4,1)
plt.imshow(r, cmap="gray")
plt.subplot(1,4,2)
plt.imshow(g, cmap="gray")
plt.subplot(1,4,3)
plt.imshow(b, cmap="gray")
plt.subplot(1,4,4)
plt.imshow(RGB)
```

```python
lower_pink = np.array([180, 0, 100])
upper_pink = np.array([255, 255, 230])
rgb_img = np.copy(img)
rgb_mask = cv2.inRange(rgb_img, lower_pink, upper_pink)
rgb_img[rgb_mask == 0] = [0, 0, 0]
plt.imshow( rgb_img)
```

# Get HSV Mask

```python
HSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h = HSV[:,:,0]
s = HSV[:,:,1]
v = HSV[:,:,2]
plt.figure(figsize=(20,15))
plt.subplot(1,4,1)
plt.imshow(h, cmap="gray")
plt.subplot(1,4,2)
plt.imshow(s, cmap="gray")
plt.subplot(1,4,3)
plt.imshow(v, cmap="gray")
plt.subplot(1,4,4)
plt.imshow(RGB)
```

```python
# pink mask
lower_hue = np.array([160, 0, 0])
upper_hue = np.array([180, 255, 255])
hsv_img = np.copy(HSV)
hsv_mask = cv2.inRange(hsv_img, lower_hue, upper_hue)
RGB_c = np.copy(RGB)
RGB_c[hsv_mask == 0] = [0, 0, 0]
plt.imshow(RGB_c) 
```

```python
# blue mask
lower_hue = np.array([90, 0, 0])
upper_hue = np.array([130, 255, 255])
hsv_img = np.copy(HSV)
hsv_mask = cv2.inRange(hsv_img, lower_hue, upper_hue)
RGB_c = np.copy(RGB)
RGB_c[hsv_mask == 0] = [0, 0, 0]
plt.imshow(RGB_c) 
```

# Get HLS Mask

```python
HLS = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
h = HLS[:,:,0]
l = HLS[:,:,1]
s = HLS[:,:,2]
plt.figure(figsize=(20,15))
plt.subplot(1,4,1)
plt.imshow(h, cmap="gray")
plt.subplot(1,4,2)
plt.imshow(l, cmap="gray")
plt.subplot(1,4,3)
plt.imshow(s, cmap="gray")
plt.subplot(1,4,4)
plt.imshow(RGB)
```

```python
lower_hue = np.array([160, 0, 0])
upper_hue = np.array([180, 255, 255])
hls_img = np.copy(HLS)
hls_mask = cv2.inRange(hls_img, lower_hue, upper_hue)
RGB_c = np.copy(RGB)
RGB_c[hls_mask == 0] = [0, 0, 0]
plt.imshow(RGB_c)
```

# apple mask 


```python
files = os.listdir("./")
for file in files:
    if file == "Red_Apple.jpg":
        img = cv2.imread(file)
        break

RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
HSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h = HSV[:,:,0]
s = HSV[:,:,1]
v = HSV[:,:,2]

plt.figure(figsize=(20,15))
plt.subplot(1,4,1)
plt.imshow(h, cmap= "gray")
plt.subplot(1,4,2)
plt.imshow(s, cmap= "gray")
plt.subplot(1,4,3)
plt.imshow(v, cmap= "gray")
plt.subplot(1,4,4)
plt.imshow(RGB)
```

```python
mask_apple = cv2.inRange(HSV, np.array([0, 100, 100]), np.array([360, 255, 255]))

RGB_c = np.copy(RGB)
RGB_c[mask_apple == 0] = [0, 0, 0]
plt.imshow(RGB_c)
```

```python
HLS = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
h = HLS[:,:,0]
l = HLS[:,:,1]
s = HLS[:,:,2]

plt.figure(figsize=(20,15))
plt.subplot(1,4,1)
plt.imshow(h, cmap= "gray")
plt.subplot(1,4,2)
plt.imshow(l, cmap= "gray")
plt.subplot(1,4,3)
plt.imshow(s, cmap= "gray")
plt.subplot(1,4,4)
plt.imshow(RGB)
```
