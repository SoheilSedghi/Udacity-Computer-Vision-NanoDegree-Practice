# Fourier Transform

Libraries


```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
```

```python
# perform a fast fourier transform and create a scaled, frequency transform image
def fourier(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # normalize the image color values from a range of [0,255] to [0,1] 
    image = image / 255.0

    f = np.fft.ifft2(image)
    fshift = np.fft.fftshift(f)
    frequency_tx = 20*np.log(np.abs(fshift))
    
    return frequency_tx
```

```python
def image_fourier(file):
    image = cv2.imread(file)
    rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    plt.subplot(1,2,1)
    plt.imshow(rgb)
    plt.subplot(1,2,2)
    plt.imshow(fourier(image))
```
