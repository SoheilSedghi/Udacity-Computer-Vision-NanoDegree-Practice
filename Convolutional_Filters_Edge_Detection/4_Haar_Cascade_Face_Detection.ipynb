# Face detection using Haar Cascade

## Libraries

```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2 
import os
```

# Read image

```python
rgb = mpimg.imread("1710797239367.jpeg")
gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap= "gray")
```

# Face detection

```python
#import classifier
dir = os.path.join("detector_architectures","haarcascade_frontalface_default.xml")
classifier = cv2.CascadeClassifier(dir)
face = classifier.detectMultiScale( gray,1.3 ,3 ) #(image, scaleFactor, minNeighbors)
len(face)
rgb_cp = np.copy(rgb)
for fa in face:
    # fa = x,y , width, height
    cv2.rectangle(rgb_cp, (fa[0],fa[1]), (fa[0]+fa[2],fa[1]+fa[3]), (255,0,0), 4 )
plt.imshow(rgb_cp)
```
