```python
import numpy as np 
import matplotlib.image as mpimg
import matplotlib.pyplot as plt 
import cv2
import os

# lets see the image in a new window
%matplotlib qt
```

```python
# dont know file names? ok i search it for you
address = os.listdir("./")
if address[0].endswith(".jpg"):
    image = mpimg.imread(address[0])

print(image.shape)
```

```python
# getting gray image
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

print(gray_image.shape)
```

```python
plt.subplot(1,2,1)
plt.imshow(gray_image,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(image)
```

```python
%matplotlib inline
# get each channel as a gray image
r = image[:, :, 0]
g = image[:, :, 1]
b = image[:, :, 2]

f, (ax1,ax2,ax3) = plt.subplots(1,3 ,figsize=(20,10))
ax1.set_title('R channel')
ax1.imshow(r, cmap='gray')
ax2.set_title('G channel')
ax2.imshow(g, cmap='gray')
ax3.set_title("B channel")
ax3.imshow(b, cmap='gray')
plt.show()
```
