# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
bgr_1 = cv2.imread("screen-shot-2017-06-30-at-4.30.11-pm.png")
bgr_2 = cv2.imread("screen-shot-2017-06-30-at-4.37.13-pm.png")

gray_1 = cv2.cvtColor(bgr_1, cv2.COLOR_BGR2GRAY)
gray_2 = cv2.cvtColor(bgr_2, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.title("image 1")
plt.imshow(gray_1, cmap = "gray")
plt.subplot(1,2,2)
plt.title("image 2")
plt.imshow(gray_2, cmap = "gray")
```

# Dilation and Erosion

```python
kernel = np.ones((5,5),np.uint8)
plt.figure( figsize= (15,10))

plt.subplot(2,2,1)
erode_1 = cv2.erode( gray_1, kernel, iterations = 1 )
plt.title("erode_1")
plt.imshow(erode_1, cmap = "gray")

plt.subplot(2,2,2)
erode_2 = cv2.erode( gray_2, kernel, iterations = 1 )
plt.title("erode_2")
plt.imshow(erode_2, cmap = "gray")

plt.subplot(2,2,3)
plt.title("dilate_1")
dilate_1 = cv2.dilate(gray_1, kernel, iterations = 1)
plt.imshow(dilate_1, cmap = "gray")

plt.subplot(2,2,4)
plt.title("dilate_2")
dilate_2 = cv2.dilate(gray_2, kernel, iterations = 1)
plt.imshow(dilate_2, cmap = "gray")
```

# Opening

```python
kernel = np.ones((8,8), np.uint8)

plt.figure(figsize=(15,10))

plt.subplot(1,2,1)
opening = cv2.morphologyEx(gray_1, cv2.MORPH_OPEN, kernel)
plt.imshow(opening, cmap = "gray")

plt.subplot(1,2,2)
closing = cv2.morphologyEx(gray_2, cv2.MORPH_CLOSE, kernel)
plt.imshow(closing, cmap= "gray")
```
