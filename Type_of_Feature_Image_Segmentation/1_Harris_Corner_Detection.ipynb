# HARRIS Corner Detection

## Libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Read Image

```python
bgr = cv2.imread("IMG_0046.JPG")
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.imshow(rgb)

plt.subplot(1,2,2)
plt.imshow(gray,cmap="gray")
```

## Harris

img - Input image. It should be grayscale and float32 type.
blockSize - It is the size of neighbourhood considered for corner detection
ksize - Aperture parameter of the Sobel derivative used.
k - Harris detector free parameter in the equation.

```python
corners = cv2.cornerHarris(gray,2,3,0.05)
corners = cv2.dilate(corners,None)
plt.figure(figsize=(15,10))
plt.imshow(corners, cmap="gray")
```

# Extract Strong Corners

```python
thresh = 0.1 * corners.max()

rgb_cp = np.copy(rgb)

for j in range(0,corners.shape[0]):
    for i in range(0,corners.shape[1]):
        # j for row and i for col
        if corners[j,i] > thresh:
            #draw circle at corners
            cv2.circle(rgb_cp,(i,j),3,(255,0,0),5)

plt.figure(figsize = (15,10))
plt.imshow(rgb_cp)
```
